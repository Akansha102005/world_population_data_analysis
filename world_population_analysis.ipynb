```python
%pip install pandas
import pandas as pd

# Load dataset
df = pd.read_csv("../projects/world_population.csv")

# Preview data
df.head()


```

```python
df.info()
df.describe()
df.columns
```

```python
df.sort_values(by='2022 Population', ascending=False).head(10)
```

```python
%pip install matplotlib
import matplotlib.pyplot as plt

years = ['1970 Population', '1980 Population', '1990 Population', '2000 Population',
         '2010 Population', '2015 Population', '2020 Population', '2022 Population']

for country in ['India', 'China', 'United States']:
    row = df[df['Country/Territory'] == country]
    plt.plot(years, row[years].values.flatten(), label=country)

plt.legend()
plt.title('Population Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

## Population Density by Continent (Boxplot)

## 

```python
%pip install seaborn
import seaborn as sns

# Sort data and get top 10
top10 = df.sort_values(by='2022 Population', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='2022 Population', y='Country/Territory', data=top10, palette='viridis')
plt.title('Top 10 Most Populous Countries in 2022')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Continent', y='Density (per km²)', data=df)
plt.title('Population Density by Continent')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='2022 Population', y='Growth Rate', hue='Continent', data=df)
plt.title('Growth Rate vs. 2022 Population by Continent')
plt.xlabel('2022 Population')
plt.ylabel('Growth Rate')
plt.xscale('log')  # optional: log scale for better readability
plt.tight_layout()
plt.show()
```
